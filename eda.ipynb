```python
df = pd.read_excel('STEAG GARGAU.xlsx', skiprows=1, converters={'Data de execução':str})
                
cols = df.select_dtypes(object).columns
df[cols] = df[cols].apply(lambda x: x.str.strip())

df_original = df.copy()

df = df[df.columns.drop(list(df.filter(regex='(?<=\.)\d+')))]
df = df[df.columns.drop(list(df.filter(regex='Unnamed')))]

df = df[df['Data de execução'].notna()]
df['observacao'] = None
```

```python
def validate(date_text):
    try:
        datetime.datetime.strptime(date_text, '%Y-%m-%d %H:%M:%S')
        return True
    except ValueError:
        return False

for index, row in df.iterrows():
  if not validate(row['Data de execução']):
      df.at[index, 'observacao'] = row['Data de execução']
      df.at[index, 'Data de execução'] = row['Iniciar']
    #   row['observacao'] = row['Data de execução']
    #   print(row.index)
    #   row['Data de execução'] = None
  
#     print(row['c1'], row['c2'])
# re.findall(r'[\d]{1,2}/[\d]{1,2}/[\d]{4}', file)[0]
# p = re.compile(r'[\d]{1,2}/[\d]{1,2}/[\d]{4}')
# df['valid'] = np.where(df['Data de execução'].str.contains(p), 1, 0)
```
